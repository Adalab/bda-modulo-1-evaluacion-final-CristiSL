agregar_producto(nombre, precio, cantidad) : Agrega un producto al inventario o
actualiza su cantidad si ya existe. Debe recibir el nombre, precio y cantidad del producto como parámetros.
Itera a través del inventario y compara los nombres de los productos con el nombre proporcionado.
Si el producto ya existe, actualiza la cantidad.
Si no existe, agrega un nuevo producto al inventario.

```python
def add_product(name, price, amount):      # Define la función que queremos crear
    for product in inventory:              # Inicia el bucle for para iterar con todo el inventario
        if product["name"] == name:        # Compara los nombres del inventario con el nombre proporcionado
            product["amount"] += amount    # Si lo encuentra, actualiza la cantidad
            return                         # Sale de la función y devuelve un valor
    inventory.append({"name": name, "price": price, "amount": amount}) # Si no lo encuentra, lo agrega al inventario

```

```python
add_product("Shirt", 20, 10)
add_product("Pants", 30, 5)
add_product("Shoes", 50, 20)
# Nuestra lista estaba vacía, así que en los 3 casos ha añadido el producto
```

buscar_producto(nombre): Busca un producto en el inventario por nombre y muestra sus
detalles si se encuentra. Debe recibir el nombre del producto como parámetro.
Utiliza un bucle for para recorrer el inventario.
Compara los nombres de los productos con el nombre proporcionado.
Si se encuentra el producto, imprime sus detalles.

```python
def search_product(name):             # Define la función que queremos crear
    for product in inventory:         # Inicia el bucle for para iterar con todo el inventario
        if product["name"] == name:   # Compara los nombres del inventario con el nombre proporcionado
            print(f"Name: {product['name']}, Price: {product['price']} €, Amount: {product['amount']}") # Si lo encuentra, imprime sus detalles
            return                    # Sale de la función y devuelve un valor
    print(f"Product not found")       # Si no lo encuentra, muestra "Product not found"
```

<span style="font-size: 18px;">Ejercicio 5:</span>

eliminar_producto(nombre): Elimina un producto del inventario por nombre. Debe recibir el
nombre del producto como parámetro.
Utiliza un bucle for para recorrer el inventario.
Busca el producto por nombre.
Elimina el producto del inventario si existe.
Si el producto no esta en el inventario muestra un mensaje indicándolo.

```python
def delete_product(name):              # Define la función que queremos crear
    for product in inventory:          # Inicia el bucle for para iterar con todo el inventario
        if product["name"] == name:    # Compara los nombres del inventario con el nombre proporcionado
            inventory.remove(product)  # Si lo encuentra, lo borra del inventario
            return                     # Sale de la función y devuelve un valor
    print(f"Product not found")        # Si no lo encuentra, muestra "Product not found"
```

<span style="font-size: 18px;">Ejercicio 6:</span>

calcular_valor_inventario(): Calcula y muestra el valor total del inventario.
Utiliza un bucle for para calcular el valor total del inventario.
Itera a través del inventario y suma el valor de cada producto (precio x cantidad). Es decir,
calcula el valor total del inventario.

```python
def calculate_inventory_value():                            # Define la función que queremos crear
    total_value = 0.0                                       # Creo la variable vacía para almacenar el resultado
    for product in inventory:                               # Inicia el bucle for para iterar con todo el inventario
        total_value += product["amount"] * product["price"] # Calcula el valor total y lo almacena en la variable
    print(f"Total_value = {total_value} €")                 # Muestra el resultado
```

<span style="font-size: 18px;">Ejercicio 7:</span>

realizar_compra(): Permite a un cliente realizar una compra seleccionando productos del
inventario. Debe interactuar con el cliente para seleccionar productos y calcular el costo total de la compra.
Utiliza un bucle while para permitir al cliente realizar múltiples compras.
Muestra el inventario y solicita al cliente ingresar el nombre del producto que desea comprar.
Registra los productos seleccionados en un carrito y actualiza el inventario.
Calcula el costo total de la compra.

```python
def make_purchase():        # Define la función que queremos crear
    cart = []               # Creo el carrito vacío para después añadirle lo que el cliente quiera
    client_purchase = 0.0   # Creo el costo total de la compra vacío porque aún no hay nada dentro
    while True:             # Inicia el bucle while para que permita hacer varias compras
        show_inventory()    # Muestra el inventario al cliente
        product_purchase = input("What product do you want to buy? or type Done if you have finished") # Pregunto al cliente qué producto quiere comprar
        if product_purchase.title() == "Done": # Si el cliente escribe "Done" sin importar mayúsculas o minúsculas
            break   # El bucle para de preguntar al cliente y sigue con el resto de pasos
        purchase_amount = input("How many do you want to buy?") # Pregunto al cliente cuántas unidades quiere
        for product in inventory: # Inicia el bucle for para iterar con todo el inventario
            if product_purchase.lower() == product["name"].lower(): # Si el producto escrito por el cliente coincide con el inventario sin importar mayúsculas o minúsculas
                cart.append({"name": product_purchase, "amount": purchase_amount}) # Añade al carrito el producto y la cantidad elegidos por el cliente
                client_purchase +=  int(purchase_amount) * product["price"] # Calcula el importe de la compra en tipo entero
                product["amount"] -= int(purchase_amount)  # Actualiza el inventario restando los producto que compra el cliente en tipo entero
                print(f"Added {product["name"]} to cart. Current total: {client_purchase} €") # Muestra al cliente lo que ha añadido al carrito 
                break # Para el bucle for para volver al bucle while
    print(f"This is your full cart:") # Muestra al cliente This is your full cart:
    for product in cart:   # Inicio un bucle for para poder mostrarle al cliente los productos que ha añadido al carrito y en qué cantidad
        print(f"Name: {product['name']}, Amount: {product['amount']}") # Muestro al cliente los productos añadidos y en qué cantidad
    print(f"Total cost: {client_purchase} €") # Muestro al cliente el precio total de su compra
```

```python
make_purchase()
```

procesar_pago(): Procesa el pago de una compra, calcula el cambio y muestra un mensaje de confirmación.
Utiliza un bloque try...except para manejar excepciones.
Solicita al cliente ingresar la cantidad total y la cantidad de pago usando un input.
Calcula el cambio y muestra un mensaje de pago exitoso o un error en caso de monto insuficiente.

```python
def process_payment():  # Define la función que queremos crear
    change = 0.0        # Creo la variable del cambio vacía para después poder calcularlo
    try:                # Inicia el bloque try para poder manejar errores
        total_payment = float(input(f"Enter the total purchase amount to pay")) # Pregunto al cliente por el pago total que debe hacer
        payment_made = float(input(f"Enter the amount you are going to pay"))   # Pregunto al cliente cuánto va a pagar él
        if payment_made < total_payment: # Si el monto de pago es menor que el monto total
            print(f"Insufficient amount. Please try again.") # Muestra un mensaje que indica que el monto es insuficiente
        else:  # Si no se cumple lo anterior
            change = payment_made - total_payment  # Calcula el cambio que le corresponde al cliente
            print(f"Payment successful, here is your change: {change} €") # Muestra un mensaje de pago exitoso e indica el cambio a devolver
    except:  # Si el cliente en el input introduce un valor no numérico
        print(f"Incorrect quantity, please try again")  # Muestra un mensaje de error 
            
```

```python
process_payment()
```

el nombre y el correo electrónico del cliente como parámetros.
Agrega un cliente al diccionario de clientes con su nombre y correo electrónico.

```python
def add_customer(name, email): # Define la función que queremos crear
    customers[name] = email    # Añado el cliente al diccionario utilizando name como clave e email como valor
    print(f"Customer '{name}' added with email '{email}'.")  # Muestra un mensaje indicando que ya lo añadió y sus datos
```
